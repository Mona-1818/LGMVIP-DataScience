```python
# importing module 
import cv2 as cv
import matplotlib.pyplot as plt 
from termcolor import colored
print(colored("lets Go!!!", "cyan", attrs=['bold']))
```

```python
def sketch(image):
    image = cv.imread(image)
    image = cv.cvtColor(image, cv.COLOR_RGB2BGR)
    grey = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    INVERT = cv.bitwise_not(grey)
    blur = cv.GaussianBlur(INVERT, (111,111),0)
    invertor = cv.bitwise_not(blur)
    sketch = cv.divide(grey, invertor, scale = 256.0)
    sketchs = cv.cvtColor(sketch, cv.COLOR_RGB2BGR)
    cv.imwrite('sketch.png', sketch)
    print(colored("Original", "magenta", attrs=['bold']))
    plt.imshow(image)
    plt.axis(True)
    plt.show()
    print("\n"*2)
    print(colored("Sketch", "magenta", attrs=['bold']))
    plt.imshow(sketchs)
    plt.axis(True)
    plt.show()
    
sketch(str(input("Enter the location: ")))
```
